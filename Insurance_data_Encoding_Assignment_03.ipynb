### One hot using Loop

```python
df7.head()
```

```python
colls=df7.columns
colls
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in colls:
    if is_numeric_dtype(df7[col]):
        continue
    one=pd.get_dummies(df7[col],drop_first=True, prefix='dummy')
    df7.drop(df7[[col]], axis=1,inplace=True)
    df7=pd.concat([df7,one],axis=1)

df7.head()
```

## OrdinalEncoder with loop

```python
df9.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
features=df9.columns
features
```

```python
for ord in features:
    if is_numeric_dtype(df9[ord]):
        continue
        
    else:
        ord_encoer=OrdinalEncoder(categories=[df9[ord].unique()])
        enc_values=ord_encoer.fit_transform(df9[[ord]])
        enc_values=pd.DataFrame(enc_values,columns=[ord])
        df9=pd.concat([df9.drop(ord,axis=1),enc_values],axis=1)
df9.head()
```
